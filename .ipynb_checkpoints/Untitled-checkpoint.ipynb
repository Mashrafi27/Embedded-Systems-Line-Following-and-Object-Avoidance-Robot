```python
def region_of_interest(image):
    height = image.shape[0]
    polygons = np.array([
        [(0, height), (0, height * 0.5), (image.shape[1], height * 0.5), (image.shape[1], height)]
    ], dtype=np.int32)
    mask = np.zeros_like(image)
    cv2.fillPoly(mask, polygons, 255)
    masked_image = cv2.bitwise_and(image, mask)
    return masked_image
```

```python
def detect_lines(image):
    # Edge detection
    edges = cv2.Canny(image, 50, 150)
    # Define ROI
    cropped_edges = region_of_interest(edges)
    # Line detection
    lines = cv2.HoughLinesP(
        cropped_edges,
        rho=1,
        theta=np.pi / 180,
        threshold=50,
        minLineLength=50,
        maxLineGap=10
    )
    return lines
```

```python
def average_slope_intercept(image, lines):
    left_fit = []
    right_fit = []
    left_weights = []
    right_weights = []
    
    for line in lines:
        for x1, y1, x2, y2 in line:
            # Fit a linear polynomial (y = mx + b)
            parameters = np.polyfit((x1, x2), (y1, y2), 1)
            slope = parameters[0]
            intercept = parameters[1]
            length = np.sqrt((y2 - y1)**2 + (x2 - x1)**2)
            if slope < 0:
                left_fit.append((slope, intercept))
                left_weights.append(length)
            else:
                right_fit.append((slope, intercept))
                right_weights.append(length)
    
    # Average using the line length as weight
    left_fit_average = np.average(left_fit, axis=0, weights=left_weights) if left_fit else None
    right_fit_average = np.average(right_fit, axis=0, weights=right_weights) if right_fit else None
    
    return left_fit_average, right_fit_average
```

```python
def make_line_points(y1, y2, line_params):
    if line_params is None:
        return None
    slope, intercept = line_params
    x1 = int((y1 - intercept) / slope)
    x2 = int((y2 - intercept) / slope)
    return np.array([x1, y1, x2, y2])
```

```python
def display_lines(image, lines, color=(0, 255, 0), thickness=5):
    line_image = np.zeros_like(image)
    if lines is not None:
        for line in lines:
            if line is not None:
                x1, y1, x2, y2 = line
                cv2.line(line_image, (x1, y1), (x2, y2), color, thickness)
    return line_image
```

```python
def process_frame(frame):
    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian Blur to reduce noise and smooth the image
    blur = cv2.GaussianBlur(gray, (5, 5), 0)

    # Binary thresholding to segment the line
    # Assuming the line is darker than the background
    _, binary = cv2.threshold(blur, 200, 255, cv2.THRESH_BINARY_INV)

    # Morphological operations to eliminate small objects and fill gaps
    kernel = np.ones((5, 5), np.uint8)

    # Closing to fill small gaps in the line
    closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

    # Opening to remove small objects or noise
    opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)

    # Edge detection using Canny Edge Detector
    edges = cv2.Canny(opening, 50, 150)

    # Optionally, define a Region of Interest (ROI) if needed
    height, width = edges.shape
    mask = np.zeros_like(edges)
    polygon = np.array([[
        (0, height),
        (0, int(height * 0.5)),
        (width, int(height * 0.5)),
        (width, height)
    ]], np.int32)
    cv2.fillPoly(mask, polygon, 255)
    cropped_edges = cv2.bitwise_and(edges, mask)

    return cropped_edges
```

```python
def process_image(image):
    # Step 1: Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Step 2: Apply Gaussian blur
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    # Step 3: Edge detection
    edges = cv2.Canny(blur, 50, 150)
    # Step 4: Define ROI and mask edges
    cropped_edges = edges # region_of_interest(edges)
    # Step 5: Line detection
    lines = cv2.HoughLinesP(
        cropped_edges,
        rho=1,
        theta=np.pi / 180,
        threshold=50,
        minLineLength=50,
        maxLineGap=10
    )
    # Step 6: Average and extrapolate lines
    left_fit, right_fit = average_slope_intercept(image, lines)
    y1 = image.shape[0]
    y2 = int(y1 * 0.6)
    left_line = make_line_points(y1, y2, left_fit)
    right_line = make_line_points(y1, y2, right_fit)
    # Step 7: Draw lines
    line_image = display_lines(image, [left_line, right_line])
    # Step 8: Overlay lines on original image
    output = cv2.addWeighted(image, 0.8, line_image, 1, 1)
    return output
```

```python
def process_frame(frame):
    # Convert the image to HSV color space
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define HSV range for black color
    # Adjust these values based on your specific lighting conditions
    lower_black = np.array([0, 0, 40])
    upper_black = np.array([180, 255, 80])

    # Create a mask for black color
    mask = cv2.inRange(hsv, lower_black, upper_black)

    # Apply morphological operations to clean up the mask
    kernel = np.ones((5, 5), np.uint8)
    
    # Closing to fill small holes in the detected regions
    closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    
    # Opening to remove small noise outside the desired regions
    opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)

    # Edge detection (optional, depending on your next processing steps)
    edges = cv2.Canny(opening, 50, 150)

    # Define a Region of Interest (ROI) if necessary
    height, width = mask.shape
    mask_roi = np.zeros_like(mask)
    polygon = np.array([[
        (0, height),
        (0, int(height * 0.2)),
        (width, int(height * 0.2)),
        (width, height)
    ]], np.int32)
    cv2.fillPoly(mask_roi, polygon, 255)
    masked_edges = cv2.bitwise_and(edges, mask_roi)

    return masked_edges
```

```python
def find_black_line_contours(masked_edges):
    _, contours, _ = cv2.findContours(masked_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    line_contours = []
    width_threshold = 100  # Adjust based on your line's pixel width

    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if w > width_threshold:
            line_contours.append(cnt)
    return line_contours
```

```python
def calculate_steering_angle(image, left_fit, right_fit):
    if left_fit is None and right_fit is None:
        return -90  # Assume a default steering angle
    elif left_fit is not None and right_fit is not None:
        # Average the positions
        mid = image.shape[1] / 2
        left_x = (left_fit[1] + (image.shape[0] - left_fit[1]) / left_fit[0])
        right_x = (right_fit[1] + (image.shape[0] - right_fit[1]) / right_fit[0])
        lane_center = (left_x + right_x) / 2
        offset = lane_center - mid
        angle = np.arctan(offset / (image.shape[0] / 2))
        angle = np.degrees(angle)
    elif left_fit is not None:
        # Steer left
        angle = 45
    else:
        # Steer right
        angle = -45
    return angle
```

```python
def simulate_robot(image):
    # Process image and get line fits
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    edges = cv2.Canny(blur, 50, 150)
    cropped_edges = region_of_interest(edges)
    lines = cv2.HoughLinesP(
        cropped_edges,
        rho=1,
        theta=np.pi / 180,
        threshold=50,
        minLineLength=50,
        maxLineGap=10
    )
    left_fit, right_fit = average_slope_intercept(image, lines)
    angle = calculate_steering_angle(image, left_fit, right_fit)
    print(f"Steering Angle: {angle} degrees")
    # Visualize
    y1 = image.shape[0]
    y2 = int(y1 * 0.6)
    left_line = make_line_points(y1, y2, left_fit)
    right_line = make_line_points(y1, y2, right_fit)
    line_image = display_lines(image, [left_line, right_line])
    output = cv2.addWeighted(image, 0.8, line_image, 1, 1)
    return output
```

```python
image = cv2.imread('./dataset/straight/0271462a-83e8-11ef-8c2e-7ac0a4e907bd.jpg')
output = simulate_robot(image)
plt.imshow(output)
plt.show()
```

```python
def detect_green_box(frame):
    # Convert to HSV color space
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define HSV range for green color
    lower_green = np.array([75, 40, 40])
    upper_green = np.array([95, 255, 255])

    # Create a mask for green color
    mask = cv2.inRange(hsv, lower_green, upper_green)

    # Find contours of the green areas
    _, contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    area_threshold = 20
    green_boxes = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > area_threshold:  # Set area_threshold appropriately
            green_boxes.append(cnt)
    return green_boxes
```
